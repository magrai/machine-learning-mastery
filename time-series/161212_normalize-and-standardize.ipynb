## 2. Normalize data

- Normalization is a rescaling of the data from the original range so that all values are within a range of 0 and 1.
- Normalization requires that you know or able to accurately estimate the minimum and maximum observable values.
- You can normalize your dataset using the scikit-learn object MinMaxScaler
- The scaler requires data to be provided as a matrix of rows and columns. When time series data is loaded as a Pandas Series, it must then be reshaped into a matrix of one column.

### Good practice usage with rescaling techniques ###
**1. Fit the scaler using available training data**
<br>The training data will be used to estimate the minimum and maximum observable values. This is done by calling the fit() function.

**2. Apply the scale to training data**
<br>You can use the normalized data to train your model. This is done by calling the transform() function.

**3. Apply the scale to data going forward**
<br>You can prepare new data in the future on which you want to make predictions.

- If needed, the transform can be inverted. This is useful for converting predictions back to their original scale for reporting or plotting. This can be done by calling the inverse_transform() function.

## Standardize

- Standardization involves rescaling the distribution of values so that the mean of observed values is 0 and the standard deviation is 1. This can be thought of as subtracting the mean or centering the data.
- Standardization assumes that your observations fit a Gaussian distribution (bell curve).
- Standardization requires that you know or able to accurately estimate the mean and standard deviation of observable values.
- You can standardize your dataset using the scikit-learn object StandardScaler.

```python
dat.hist()
plt.show()
```

```python
## Print first 5 rows
for i in range(5):
    print(values[i])

## Train the standardization
scaler_standardize = StandardScaler()
scaler_standardize = scaler_standardize.fit(values)
print("---")
print('Mean: %f' % (scaler_standardize.mean_))
print('StandardDeviation: %f' % (sqrt(scaler_standardize.var_)))

## Standarize the dataset
values_standardized = scaler_standardize.transform(values)

## Print first 5 rows
print("---")
for i in range(5):
    print(values_standardized[i])

## Inverse transform
values_standardized_inversed = scaler_standardize.inverse_transform(values_standardized)

## Print first 5 rows
print("---")
for i in range(5):
    print(values_standardized_inversed[i])
```
